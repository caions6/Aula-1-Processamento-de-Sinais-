1. Para um sinal x(t) = cos(2πf t) com dura¸c˜ao de 5 s e frequˆencia de amostragem fs = 44.1 kHz, fa¸ca o que se
pede nos itens abaixo:


(a) Gere gr´aficos de x(t) no dom´ınio do tempo para f ∈ {500, 5000, 10000} Hz.

(b) Escreva um c´odigo para tocar os sinais gerados.

(c) Comente os resultados obtidos.

```python
import numpy as np # Importa a biblioteca NumPy para operações numéricas, especialmente para arrays.
import matplotlib.pyplot as plt # Importa a biblioteca Matplotlib para criar gráficos e visualizações.

# Parâmetros do sinal
duration = 5  # Duração do sinal em segundos
fs = 44100    # Frequência de amostragem em Hertz (amostras por segundo)
frequencies = [500, 5000, 10000] # Frequências dos sinais em Hertz a serem plotados

# Vetor de tempo
# Cria um array de tempo que vai de 0 até 'duration' (exclusivo), com passos de 1/fs.
# Isso garante que temos 'fs' amostras por segundo.
t = np.arange(0, duration, 1/fs)

# Gera e plota x(t) para cada frequência
# Cria uma figura para os gráficos com um tamanho específico para melhor visualização.
plt.figure(figsize=(15, 10))

# Itera sobre cada frequência na lista 'frequencies'
for i, f in enumerate(frequencies):
    # Gera o sinal cosseno x(t) = cos(2 * pi * f * t)
    x_t = np.cos(2 * np.pi * f * t)

    # Cria um subplot para cada frequência.
    # len(frequencies) é o número total de linhas, 1 coluna, e i + 1 é a posição atual do subplot.
    plt.subplot(len(frequencies), 1, i + 1)
    # Plota o sinal x_t em função do tempo t
    plt.plot(t, x_t)
    # Define o título do subplot, indicando a frequência atual.
    plt.title(f'x(t) para f = {f} Hz')
    # Define o rótulo do eixo X como 'Tempo (s)'
    plt.xlabel('Tempo (s)')
    # Define o rótulo do eixo Y como 'Amplitude'
    plt.ylabel('Amplitude')
    # Adiciona uma grade ao gráfico para facilitar a leitura dos valores.
    plt.grid(True)
    # Limita o eixo X para um pequeno intervalo (0 a 0.01 segundos).
    # Isso é feito porque a duração total de 5s é muito longa para visualizar as oscilações individuais do sinal.
    plt.xlim(0, 0.01)

# Ajusta o layout para evitar sobreposição de títulos e rótulos.
plt.tight_layout()
# Exibe todos os gráficos gerados.
plt.show()
```

```python
# (b) Escreva um código para tocar os sinais gerados.

from IPython.display import Audio # Importa a função Audio para tocar sons diretamente no Colab
import scipy.io.wavfile as wavfile # Importa a biblioteca para salvar arquivos WAV

# Certifica-se de que as variáveis do sinal (t, fs, frequencies) estão disponíveis do item (a)
# Se este código fosse executado sozinho, as variáveis precisariam ser definidas novamente.
# Por exemplo:
# import numpy as np
# duration = 5  # Duração do sinal em segundos
# fs = 44100    # Frequência de amostragem em Hertz (amostras por segundo)
# frequencies = [500, 5000, 10000] # Frequências dos sinais em Hertz
# t = np.arange(0, duration, 1/fs)

print("Gerando e tocando os sinais de áudio...")

for f in frequencies:
    # Gera o sinal cosseno x(t) = cos(2 * pi * f * t)
    # Usamos o array 't' que já foi definido anteriormente para toda a duração.
    x_t_full = np.cos(2 * np.pi * f * t)

    # Normalização do sinal para garantir que o áudio não 'estoure' e esteja em um formato adequado.
    # Convertemos para formato de inteiro de 16 bits, que é comum para áudio.
    # Multiplicamos pelo valor máximo de um int16 (32767) para escalar o sinal de -1 a 1 para -32767 a 32767.
    audio_data = (x_t_full * 32767).astype(np.int16)

    # Define o nome do arquivo para salvar o áudio
    filename = f'signal_{f}Hz.wav'

    # Salva o sinal como um arquivo WAV
    # O primeiro argumento é a frequência de amostragem (fs), o segundo são os dados de áudio.
    wavfile.write(filename, fs, audio_data)

    print(f"Sinal de {f} Hz salvo como {filename} e pronto para tocar:")
    # Exibe o player de áudio no Colab para o arquivo gerado
    display(Audio(filename))
    print("\n") # Adiciona uma linha em branco para melhor separação visual

print("Concluído a geração e exibição dos players de áudio.")
```
